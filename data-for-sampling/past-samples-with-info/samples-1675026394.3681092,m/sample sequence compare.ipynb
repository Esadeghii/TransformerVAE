```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_sample= pd.read_csv('detailed-sequences.csv')
data_train= pd.read_csv('data-info-model-4.csv')
data_reencode= pd.read_csv('reencoded-data-info-model-4.csv')
```

```python
overlaps=[]
for i in data_train['Sequence']:
    if i in data_reencode['Sequence']:
        overlaps.append(i)
        
```

```python
overlaps
```

```python
data_reencode['Sequence']== data_sample['Sequence Generated']
```

```python
plt.hist(data_train['Z-Wavelength'], bins= 22, rwidth= 0.8, color= 'k', label='train')
plt.hist(data_sample['Wavelen Generated'], bins= 22, rwidth= 0.5, color= 'b', label='sample')
plt.xlabel('WAV proxy')
plt.ylabel('probability density')
plt.title('WAV Proxy samples')
plt.legend()
plt.show()
```

```python
plt.hist(data_train['Z-Wavelength'], bins= 22, rwidth= 0.8, color= 'k',label='train')
plt.hist(data_reencode['Z-Wavelength'], bins= 22, rwidth= 0.5, color= 'r', label='Re-encoded')
plt.xlabel('WAV proxy')
plt.ylabel('probability density')
plt.title('WAV Proxy re-encoded')
plt.legend()
plt.show()
```

```python
plt.hist(data_train['Z-LII'], bins= 22, rwidth= 0.8, color= 'k',label='train')
plt.hist(data_sample['LII Generated'], bins= 22, rwidth= 0.5, color= 'b', label='sample')
plt.xlabel('lii proxy')
plt.ylabel('probability density')
plt.title('lii Proxy samples')
plt.legend()
plt.show()
```

```python
plt.hist(data_train['Z-LII'], bins= 22, rwidth= 0.8, color= 'k', label='train')
plt.hist(data_reencode['Z-LII'], bins= 22, rwidth= 0.5, color= 'r', label='Re-encoded')
plt.xlabel('LII proxy')
plt.ylabel('probability density')
plt.title('LII Proxy re-encoded')
plt.legend()
plt.show()
```
